```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
import matplotlib as mpl
import matplotlib.pyplot as plt
from scipy import stats
from numpy.random import randn
import seaborn as sns
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.lmplot('total_bill','tip',tips)
```

```python
sns.lmplot('total_bill','tip',tips,order=4,scatter_kws={'marker':'o','color':'indianred'},line_kws={'linewidth':1,'color':'blue'})
```

```python
sns.lmplot('total_bill','tip',tips,fit_reg=False)
```

```python
tips['tip_percent']=100*(tips['tip']/tips['total_bill'])
```

```python
tips
```

```python
sns.lmplot('size','tip_percent',tips)
```

```python
sns.lmplot('size','tip',tips,x_jitter=1)
```

```python
sns.lmplot('total_bill','tip_percent',tips,hue='sex',markers=['x','o'])
```

```python
sns.lmplot('total_bill','tip_percent',tips,hue='day')
```

```python
sns.lmplot('total_bill','tip_percent',tips,lowess=True,line_kws={'color':'black'})
```

```python
sns.regplot('total_bill','tip_percent',tips)
```

```python
fig,(axis1,axis2)=plt.subplots(1,2,sharey=True)
sns.regplot('total_bill','tip_percent',tips,ax=axis1)
sns.violinplot(tips['tip_percent'],tips['size'],color='indianred',ax=axis2)
```
