```python
import numpy as np
import pandas as pd
from scipy import stats
from numpy.random import randn
from pandas import Series, DataFrame
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
%matplotlib inline
```

```python
data1= randn(100)
data2= randn(100)
```

```python
sns.boxplot([data1,data2])
```

```python
sns.boxplot([data1,data2],whis=np.inf)
```

```python
sns.boxplot(data2)
```

```python
sns.boxplot(data1,color='indianred',vert=True)
```

```python
data1=stats.norm(0,5).rvs(100)
data2=np.concatenate([stats.gamma(5).rvs(50)-1,stats.gamma(5).rvs(50)*-1])
sns.boxplot([data1,data2],whis=np.inf)
```

```python
sns.boxplot(data2)
```

```python
sns.boxplot(data1,color='indianred')
```

```python
sns.violinplot([data1,data2])
```

```python
sns.violinplot(data2)
```

```python
sns.violinplot(data1,bw=0.01)
```

```python
sns.violinplot(data1,inner='stick')
```
