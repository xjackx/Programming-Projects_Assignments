```python
import numpy as np
import pandas as pd
from pandas import Series,DataFrame
import datetime as dt
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
na_data=np.loadtxt('D:\Ebooks\MSc Big Data -Stirling University\Online Courses_MOOCS\Computational Investing Part I\QSTK\example.csv',delimiter=',',skiprows=1)
na_price=na_data[:,3:]
na_dates=np.int_(na_data[:,0:3])
ls_symbols=['$SPX','XOM','GOOG','GLD']
```

```python
print "First 5 rows of Price Data:"
print na_price[:5,:]
print "First 5 rows of Dates:"
print na_dates[0:5,:]
```

```python
ldt_timestamps=[]
for i in range(0,na_dates.shape[0]):
    ldt_timestamps.append(dt.date(na_dates[i,0],na_dates[i,1],na_dates[i,2]))
```

```python
plt.clf()
plt.plot(ldt_timestamps,na_price[:,:]/na_price[0,:])
plt.legend(ls_symbols)
plt.ylabel('Adjusted Close')
plt.xlabel('Date')
plt.savefig('adjustedclose.pdf',format='pdf')
```
